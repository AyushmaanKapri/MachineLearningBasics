```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]
plt.plot(x, y)
```

```python
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Simple Line Plot')
plt.plot(x,y,color='green',linestyle='--',marker='o',linewidth=2)
plt.grid(True)
```

```python
## Multiple Plots
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 13, 17, 20]
y2 = [12, 14, 11, 19, 22]

plt.figure(figsize=(10, 5))
plt.title('Plot 1')
plt.xlabel('x-axis')
plt.ylabel('y1-axis')

plt.subplot(1, 2, 2)
plt.plot(x, y1, color='blue', marker='o')

plt.subplot(1, 2, 1)
plt.plot(x, y2, color='red', marker='x')

plt.subplot(1, 2, 3)
plt.plot(x, y1, color='blue', marker='o')

plt.subplot(1, 2, 4)
plt.plot(x, y2, color='red', marker='x')

```
